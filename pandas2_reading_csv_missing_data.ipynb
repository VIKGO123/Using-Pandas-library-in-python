```python
import pandas as pd
df=pd.read_csv("C:/Users/hp/Desktop/weather.csv",parse_dates=["day"])#using dates as index converts date to standard form
df.set_index('day',inplace=True)# inplace is used to put day inplace of index
new_df=df.fillna({
    'temp':0,'windspeed':0,'event':'Not sure'# using dictionary in fillna
})
new_df
```

```python
new_df=df.fillna(method="ffill")#ffill is forward fill=fill with previous value
new_df#use tab at . to get list of methods that can be used
```

```python
new_df=df.fillna(method="bfill")#bfill is backward fill-fill with coming value
new_df
new_df.describe()
```

```python
new_df=df.fillna(method="bfill",axis="columns")#bfill is backward fill-fill with coming value :axis fills data horizontally
new_df
#new_df.describe()
```

```python
new_df=df.fillna(method="ffill",limit=2)# limit fills the no.of empty entries below a complete entry
new_df
```

```python
new_df=df.interpolate()#interpolates two data for a missing between data
new_df
```

```python
new_df=df.dropna()#exits fields having atleast one empty column
new_df
```

```python
new_df=df.dropna(how="all")#exits fields having all empty columns 
new_df
```

```python
dt=pd.date_range("01-01-2017","01-11-2017")#fills in empty places in dates
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```
